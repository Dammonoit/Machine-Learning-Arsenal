# **Matplotlib.**

* matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB.
* blue line is by-default in matplotlib and blue color also is by-default.
    * The default format string is 'b-', which is a solid blue line 
* **"%matplotlib inline"** syntax has a paradoxical effect because sometimes your plot may print and sometimes it'll not so it's safe to add it.
* pyplot is a module that collects a couple of functions that allow matplotlib to be used in a functional manner. 
* The pyplot is imported as **import matplotlib.pyplot as plt**. In this case, there are three different commands that remove stuff:
    * **plt.cla()** clears an axes, i.e. the currently active axes in the current figure. It leaves the other axes untouched.
    * **plt.clf()** clears the entire current figure with all its axes, but leaves the window opened, such that it may be reused for other plots.
    * **plt.close()** closes a window, which will be the current window, if not specified otherwise.

## **with categorical variables.**
* It is also possible to create a plot using categorical variables. Matplotlib allows you to pass categorical variables directly to many plotting functions i.e types of plots as in bar,histogram etc
* The subplot() command specifies numrows, numcols, plot_number where plot_number ranges from 1 to numrows*numcols. The commas in the subplot command are optional if numrows*numcols<10. So subplot(211) is identical to subplot(2, 1, 1).

```python
#this is also an example of multiple figures.
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)   #bar chart
plt.subplot(132)
plt.scatter(names, values)  #scatter chart
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')    #suptitle means supertitle i.e title of various figures one at a time.
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot2grid((3,3), (0,0))     #it's a grid format and (3,3),(0,0) is the location.
plt.bar(names, values)
plt.subplot2grid((3,3), (1,0)) 
plt.scatter(names, values)
plt.subplot2grid((3,3), (2,0)) 
plt.plot(names, values)
plt.suptitle('Categorical Plotting')    #suptitle means supertitle i.e title of various figures one at a time.
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot2grid((3,3), (0,0)) 
plt.bar(names, values)
plt.subplot2grid((3,3), (1,0)) 
plt.scatter(names, values)
plt.subplot2grid((3,3), (0,2)) 
plt.plot(names, values)
plt.suptitle('Categorical Plotting')    #suptitle means supertitle i.e title of various figures one at a time.
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot2grid((3,3), (0,0)) 
plt.bar(names, values)
plt.subplot2grid((3,3), (1,0)) 
plt.scatter(names, values)
plt.subplot2grid((3,3), (0,2),rowspan=3, colspan=2)  #again formatting is done to size of the last plot.
plt.plot(names, values)
plt.suptitle('Categorical Plotting')    #suptitle means supertitle i.e title of various figures one at a time.
plt.show()
```

### Legends and grids in matplotlib
* legends is the sort of naming convention where each graph is named in case of multiple graphs.
* let's see it with one example.

```python
days=[1,2,3,4,5,6,7]   #days of the week
min_t=[40,39,42,36,38,46,44]  #min temperature in all days of week.
max_t=[50,53,49,51,58,45,51]  #max temp.
avg_t=[44,43,39,41,48,45,51]  #average temp.
plt.plot(days,min_t,"r-",label="min temperature")
plt.plot(days,max_t,"b-",label="max temperature")
plt.plot(days,avg_t,"c-",label="avg temperature")
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Description')
plt.legend(loc='best',shadow='True',fontsize='large')         #it gives most optimal location as it tries to find only blank location.
plt.grid()                                      #it adds grid.
plt.show()
```

```python
days=[1,2,3,4,5,6,7]   #days of the week
min_t=[40,39,42,36,38,46,44]  #min temperature in all days of week.
max_t=[50,53,49,51,58,45,51]  #max temp.
avg_t=[44,43,39,41,48,45,51]  #average temp.
plt.plot(days,min_t,"r-",label="min temperature")
plt.plot(days,max_t,"b-",label="max temperature")
plt.plot(days,avg_t,"c-",label="avg temperature")
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Description')
plt.legend(loc='lower right',shadow='false',fontsize='small')    #it gives the legend to be lower left location.
plt.grid()
plt.show()
```

## **Working with multiple Plots and axes.**
* The figure() command in the code below is optional because figure(1) will be created by default, just as a subplot(111) will be created by default if you don't manually specify any axes.
* You can clear the current figure with clf() and the current axes with cla(). If you find it annoying that states (specifically the current image, figure and axes) are being maintained for you behind the scenes, don't despair.
* If you are making lots of figures, you need to be aware of one more thing: the memory required for a figure is not completely released until the figure is explicitly closed with close(). Deleting all references to the figure, and/or using the window manager to kill the window in which the figure appears on the screen, is not enough, because pyplot maintains internal references until close() is called.

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default


plt.title('Easy as 1, 2, 3') # subplot 211 title
```

## **Working with text.**
* The text() command can be used to add text in an arbitrary location, and the xlabel(), ylabel() and title() are used to add text in the indicated locations.
* Matplotlib has extensive text support, including support for mathematical expressions, truetype support for raster and vector outputs, newline separated text with arbitrary rotations, and unicode support.

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')  #60 and 0.25 are the labels i.e axis values.
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
#matplotlib accepts TeX equation expressions in any text expression. 
#For example to write the expression σi=15 in the title, you can write a TeX expression surrounded by dollar signs:
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\sigma_i=15$')           #writing sigma_i to the title of the graph.
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')  #60 and 0.25 are the labels i.e axis values.
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('my data', fontsize=14, color='red')   #customization on x-label.
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')  #60 and 0.25 are the labels i.e axis values.
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt

fig = plt.figure()                                #figure function is full figure.
fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)                         #a new subplot is added into the figure.
fig.subplots_adjust(top=0.85)
ax.set_title('axes title')

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')

ax.text(3, 8, 'boxed italics text in data coords', style='italic',
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})             #here (3,8) specifies the location.

ax.text(2, 6, r'an equation: $E=mc^2$', fontsize=15)

ax.text(3, 2, 'unicode: Institut für Festkörperphysik')

ax.text(0.95, 0.01, 'colored text in axes coords',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='green', fontsize=15)


ax.plot([2], [1], 'ro')                                #here r is red color and anotation is at (2,1) but text is at (3,4)
ax.annotate('annotate', xy=(2, 1), xytext=(3, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.axis([0, 10, 0, 10])

plt.show()
```

* A common use for text is to annotate some feature of the plot, and the annotate() method provides helper functionality to make annotations easy. 
* In an annotation, there are two points to consider: the location being annotated represented by the argument xy and the location of the text xytext. Both of these arguments are (x,y) tuples.

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),               #another example of text anotation.
             )

plt.ylim(-2, 2)
plt.show()
```

```python
#we can also specify the x and y labels after figure has been created via. set_xlabel() and set_ylabel()
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.0, 5.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.25)     #this is used to adjust subplot in the figure i.e bottom size and left expansion
ax.plot(x1, y1)
ax.set_xlabel('time [s]')
ax.set_ylabel('Damped oscillation [V]')

plt.show()
```
